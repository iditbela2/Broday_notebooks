```python
# Bartlett’s Test - p-value should be significant
```

```python
chi_square_value,p_value=calculate_bartlett_sphericity(df)
chi_square_value, p_value
```

```python
# Kaiser-Meyer-Olkin (KMO) Test - between 0 and 1 - when 1 is highly suitable for factor analysis
```

```python
kmo_all,kmo_model=calculate_kmo(df)
kmo_model
```

```python
fa = FactorAnalyzer()
```

```python
fa.fit(df)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
ev
```

```python
# Create scree plot using matplotlib
plt.scatter(range(1,df.shape[1]+1),ev)
plt.plot(range(1,df.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Number of factors')
plt.ylabel('Eigenvalue')
plt.axhline(y=1,c='r')
plt.show();
```

```python
NUM = 4
```

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer(n_factors=NUM, rotation='promax') #oblimin, quartimin, promax, varimax
fa.fit(df)
```

```python
fa.loadings_
```

```python
fa.get_factor_variance()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
c = ax.pcolor(fa.loadings_)
fig.colorbar(c, ax=ax)
ax.set_yticks(np.arange(fa.loadings_.shape[0])+0.5, minor=False)
ax.set_xticks(np.arange(fa.loadings_.shape[1])+0.5, minor=False)
ax.set_yticklabels(df.columns)
ax.set_xticklabels(['Factor '+str(i+1) for i in range(fa.loadings_.shape[1])])
plt.show();
```
